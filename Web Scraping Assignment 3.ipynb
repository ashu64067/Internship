```python
private = []
dorm = []
hostel_name = []
distance = []
facilities = []
prop_desc = []
rating = []
```

```python
private_tags = driver.find_elements(By.XPATH, '//div[@class="prices-col"]/a/div[1]')
for i in private_tags:
    private.append(i.text)
dorm_tags = driver.find_elements(By.XPATH, '//div[@class="prices-col"]/a/div[2]')
for i in dorm_tags:
    dorm.append(i.text)
hostel_name_tags = driver.find_elements(By.XPATH, '//div[@class="property-card"]/div[3]/div[1]/h2')
for i in hostel_name_tags:
    hostel_name.append(i.text)
distance_tags = driver.find_elements(By.XPATH , '//div[@class="property-card"]/div[3]/div[1]/div[1]/a/span/span')
for i in distance_tags:
    distance.append(i.text.split('- ')[1])
facilities_tags = driver.find_elements(By.XPATH , '//div[@class="property-card"]/div[3]/div[1]/div[2]')
for i in facilities_tags:
    if i.text == '':
        facilities.append('-')
    else:
        facilities.append(i.text)
prop_tags = driver.find_elements(By.XPATH, '//div[@class="property-card"]/div[3]/div[2]/div[1]')
for i in prop_tags:
    prop_desc.append(i.text)
rating_tags = driver.find_elements(By.XPATH , '//div[@class="property-card"]/div[3]/div[2]/div[2]/a/div')
for i in rating_tags:
    rating.append(i.text)
```

```python
hotels_df = pd.DataFrame({'Hostel Name': hostel_name,
                         'Distance from city centre': distance,
                         'Rating & Reviews': rating,
                         'Privates': private,
                         'Dorms': dorm,
                         'Facilities': facilities,
                         'Property Description': prop_desc})
```
