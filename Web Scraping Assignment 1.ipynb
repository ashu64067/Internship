```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

# Question 1

```python
def get_wikiheader(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.content)
    header = []
    for i in soup.find_all(['h1','h2']):
        header.append(i)
    wikiheaders = pd.DataFrame({'Headers List': header})
    return wikiheaders
```

```python
get_wikiheader('https://en.wikipedia.org/wiki/Main_Page')
```

# Question 2

```python
def get_moviesdata(url):
    page = requests.get(url)
    movies = BeautifulSoup(page.content)
    y = movies.find_all('h3', class_='lister-item-header')
    r = movies.find_all('div', class_='ipl-rating-star small')
    name = []
    rating = []
    year = []
    for i in y:
        name.append(i.text.split('\n')[2])
        year.append(i.text.split('\n')[3])
    for m in r:
        rating.append(m.text.split('\n')[8])
    moviesdata = pd.DataFrame({'Name': name, 'Rating': rating, 'Year': year}, index= range(1,51))
    return moviesdata
get_moviesdata('https://www.imdb.com/list/ls055386972/')
```

# Question 4

```python
def president_data(url):
    page = requests.get(url)
    pr = BeautifulSoup(page.content)
    pname = []
    tof = []
    for i in pr.find_all('h3'):
        pname.append(i.text.split('(')[0])
    for n in pr.find_all('div', class_='presidentListing'):
        tof.append(n.text.split('\n')[2].split(':')[1])
    president = pd.DataFrame({'Name': pname, 'Term of Office': tof})
    return president
president_data('https://presidentofindia.nic.in/former-presidents.htm')
```

# Question 5

Part A

```python
def cricket_info(url):
    page = requests.get(url)
    cricket = BeautifulSoup(page.content)
    name = []
    matches = []
    points = []
    rating = []
    matches.append(cricket.find('td', class_='rankings-block__banner--matches').text)
    points.append(cricket.find('td', class_='rankings-block__banner--points').text)
    rating.append(cricket.find('td', class_='rankings-block__banner--rating u-text-right').text.split()[0])
    for i in cricket.find_all('span', class_='u-hide-phablet'):
        name.append(i.text)
    for i in cricket.find_all('tr', class_='table-body'):
        matches.append(i.text.split('\n')[7])
        points.append(i.text.split('\n')[8])
        rating.append(i.text.split('\n')[9])
    cricketdata = pd.DataFrame({'Name of the team': name[0:10],'Matches': matches[0:10], 'Points': points[0:10], 'Rating': rating[0:10]}, index = range(1,11))
    return cricketdata
cricket_info('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

Part B

```python
def batter_info(url):
    page = requests.get(url)
    cricket = BeautifulSoup(page.content)
    rname = []
    rteam = []
    rRating = []
    name = []
    team = []
    rating = []
    for i in cricket.find_all('div', class_='rankings-block__banner--name'):
        rname.append(i.text)
    for i in cricket.find_all('div', class_='rankings-block__banner--nationality'):
        rteam.append(i.text.split()[0])
        rRating.append(i.text.split()[1])
    name.append(rname[0])
    team.append(rteam[0])
    rating.append(rRating[0])
    for i in cricket.find_all('td', class_='table-body__cell name'):
        name.append(i.text.split('\n')[1])
    for i in cricket.find_all('span', class_='table-body__logo-text'):
        team.append(i.text)
    for i in cricket.find_all('td', class_='table-body__cell u-text-right rating'):
        rating.append(i.text)
    batterdata = pd.DataFrame({'Name of the batter': name[0:10],'Team': team[0:10], 'Rating': rating[0:10]}, index = range(1,11))
    return batterdata
batter_info('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
```

# Question 7

```python
def news_detail(url):
    page = requests.get(url)
    news = BeautifulSoup(page.content)
    news_headline = []
    news_link = []
    timestamp = []
    for i in news.find_all('a', class_='LatestNews-headline'):
        news_headline.append(i.text)
        news_link.append(i.get('href'))
    for i in news.find_all('time', class_='LatestNews-timestamp'):
        timestamp.append(i.text)
    details = pd.DataFrame({'Headline': news_headline, 'Time': timestamp, 'News Link': news_link})
    return details
news_detail('https://www.cnbc.com/world/?region=world')
```

```python
def articles(url):
    page = requests.get(url)
    book = BeautifulSoup(page.content)
    title = []
    link = []
    author = []
    date = []
    for i in book.find_all('a', class_='sc-5smygv-0 fIXTHm'):
        title.append(i.text)
        link.append(i.get('href'))
    for i in book.find_all('span', class_='sc-1w3fpd7-0 dnCnAO'):
        author.append(i.text)
    for i in book.find_all('span', class_='sc-1thf9ly-2 dvggWt'):
        date.append(i.text)
    book_info = pd.DataFrame({'Paper Title': title, 'Authors': author, 'Published Date': date, 'Paper URL': link})
    return book_info
articles('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
```

```python
def resturant_info(url):
    page = requests.get(url)
    res = BeautifulSoup(page.content)
    name = []
    cuisine = []
    location = []
    rating = []
    images_url = []
    for i in res.find_all('a', class_="restnt-name ellipsis"):
        name.append(i.text)
    for i in res.find_all('div', class_="detail-info"):
        cuisine.append(i.text.split('|')[1])
    for i in res.find_all('div', class_="restnt-loc ellipsis"):
        location.append(i.text)
    for i in res.find_all('div', class_='restnt-rating rating-4'):
        rating.append(i.text)
    for i in res.find_all('img', class_="no-img"):
        images_url.append(i.get('data-src'))
    res_data = pd.DataFrame({'Restaurant name': name, 'Cuisine': cuisine, 'Location': location, 'Ratings': rating, 'Image URL': images_url})
    return res_data
resturant_info('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```
