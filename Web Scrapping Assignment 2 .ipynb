```python
driver.get('http://www.flipkart.com/')
```

```python
try:
    driver.find_element(By.XPATH, '//button[@class="_2KpZ6l _2doB4z"]').click()
except:
    pass
```

```python
search = driver.find_element(By.CLASS_NAME, "_3704LK")
search.send_keys('sunglasses')
```

```python
button = driver.find_element(By.XPATH, "/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
button.click()
```

```python
brand=[]
description=[]
price=[]
discount=[]
```

```python
for page in range(0,3):
    brand_tags= driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]') #This will get us all the brand name
    for i in brand_tags:
        brand.append(i.text)
    #this below code get us full description
    description_tags= driver.find_elements(By.CLASS_NAME, 'IRpwTa') 
    for i in description_tags:
        description.append(i.get_attribute('title'))
    #this will scrape prices
    price_tags= driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for i in price_tags:
        price.append(i.text)
    #this will scrape discount
    discount_tags= driver.find_elements(By.XPATH, '//div[@class="_3Ay6Sb"]')
    for i in discount_tags:
        discount.append(i.text)
    #this code will take us to next page
    button = driver.find_element(By.XPATH, "//span[text()='Next']")
    button.click()
    time.sleep(3) #this will execute next loop after 3 seconds
```

```python
sunglasses_df = pd.DataFrame({'Brand': brand[0:100], 'Product Description': description[0:100], 'Discounted Price': price[0:100], 'Discount': discount[0:100]}, index = range(1,101))
sunglasses_df
```

```python
driver.get('https://www.amazon.in/')
```

```python
search = driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search.send_keys('Laptop')
```

```python
#this will click on search button.
search_button = driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search_button.click()
```

```python
#this will filter i7 laptops
driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[6]/ul[7]/span[13]/li/span/a/div').click()
```

```python
lap_name=[]
lap_rating=[]
lap_price=[]
```

```python
#this will get whole data for laptop
all_products = driver.find_elements(By.XPATH, '//div[@data-component-type="s-search-result"]')
```

```python
for product in all_products:
    #.// means in this path that will these xpaths from the xpath from all products only.
    name_tags = product.find_elements(By.XPATH, './/span[@class="a-size-medium a-color-base a-text-normal"]')
    for name in name_tags:
        lap_name.append(name.text)
    price_tags = product.find_elements(By.XPATH, './/span[@class="a-price-whole"]')
    for price in price_tags:
        lap_price.append(price.text)
    #there are some laptops which doesn't have reviews so this below code will fill the empty reviews 
    
    if len(product.find_elements(By.XPATH, './/span[@class="a-size-base s-underline-text"]'))>0:
            rating_tags = product.find_elements(By.XPATH, './/span[@class="a-size-base s-underline-text"]')
            for rating in rating_tags:
                lap_rating.append(rating.text)
    else:
        lap_rating.append('0')
```

```python
laptop_DF = pd.DataFrame({'Laptop name': lap_name[0:10], 'Ratings': lap_rating[0:10], 'Price': lap_price[0:10]}, index=range(1,11))
```

```python
laptop_DF
```

# Question 8

```python
driver.get('https://www.azquotes.com/')
```

```python
driver.find_element(By.XPATH, '//a[@id="showhide"]').click()
top_quotes = driver.find_element(By.XPATH, '//a[@href="/top_quotes.html"]')
top_quotes.click()
```

```python
quo = []
author = []
type_quotes = []
```

```python
page = 0
while page<10:

    quo_tags = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quo_tags:
        quo.append(i.text)

    author_tags = driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in author_tags:
        author.append(i.text.split('\n')[1].split('    ')[-1])

    type_tags = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in type_tags:
        type_quotes.append(i.text.split('\n')[1].split('          ')[4])
    page += 1
    if page == 10:
        break
    
    button = driver.find_element(By.XPATH, '//li[@class="next"]')
    button.click()
    time.sleep(2)
```

```python
top_quotes_DF = pd.DataFrame({'Quote': quo, 'Author Name': author, 'Type of Quote': type_quotes}, index=range(1,1001))
```

```python
top_quotes_DF
```
