# Question 4

```python
driver.get('https://github.com/')
```

```python
try:
    driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[1]/div[2]/button').click()
except:
    pass
```

```python
driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button').click()
driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a').click()
```

```python
title=[]
desc=[]
lan=[]
```

```python
cont=[]
```

```python
for i in driver.find_elements(By.XPATH, '//article[@class="Box-row"]/h2'):
    title.append(i.text)

box = driver.find_elements(By.XPATH, '//article[@class="Box-row"]')
for n in box:
    if len(n.find_elements(By.XPATH, './/p[@class="col-9 color-fg-muted my-1 pr-4"]'))>0:
        desc_tag = n.find_elements(By.XPATH, './/p[@class="col-9 color-fg-muted my-1 pr-4"]')
        for i in desc_tag:
            desc.append(i.text)
    else:
        desc.append('-')

for n in box:
    if len(n.find_elements(By.XPATH, './/span[@itemprop="programmingLanguage"]'))>0:
        lan_tag = n.find_elements(By.XPATH, './/span[@itemprop="programmingLanguage"]')
        for i in lan_tag:
            lan.append(i.text)
    else:
        lan.append('-')
```

```python
url=[]
for i in driver.find_elements(By.XPATH, '//article[@class="Box-row"]/h2/a'):
    url.append(i.get_attribute('href'))
```

```python
for i in url:
    driver.get(i)
    time.sleep(3)
    try:
        driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/main/div/nav/div/details/summary').click()
    except:
        pass
    driver.find_element(By.XPATH, '//a[@id="insights-tab"]').click()
    time.sleep(3)
    driver.find_element(By.XPATH, '//nav[@class="menu"]/a[2]').click()
    time.sleep(10)
    a=0
    for i in driver.find_elements(By.XPATH, '//span[@class="d-block Box"]'):
        a+=1
    cont.append(a)
    time.sleep(3)
```

```python
cont_tags = driver.find_elements(By.XPATH, '//span[@class="d-inline-block mr-3"]')
for con in cont_tags:
    a=0
    for i in con.find_elements(By.XPATH, './/a[@class="d-inline-block"]'):
        a+=1
    cont.append(a)
```

```python
github_df = pd.DataFrame({'Repository title': title, 'Repository description': desc, 'Contributors count': cont,
                          'Language used': lan})
```

```python
github_df
```

# Question 8

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
driver.find_element(By.XPATH, '/html/body/table[2]/tbody/tr/td/span/b/a').click()
```

```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
instances=[]
attribute=[]
year=[]
```

```python
for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td/table/tbody/tr/td[2]'):
    dataset_name.append(i.text)

for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td[2]/p')[1:]:
    if i.text == ' ':
        data_type.append('-')
    else:
        data_type.append(i.text)
        
for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td[3]/p')[1:]:
    if i.text == ' ':
        task.append('-')
    else:
        task.append(i.text)
        
for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td[4]/p')[1:]:
    if i.text == ' ':
        attribute_type.append('-')
    else:
        attribute_type.append(i.text)
        
for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td[5]/p')[1:]:
    if i.text == ' ':
        instances.append('-')
    else:
        instances.append(i.text)
        
for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td[6]/p')[1:]:
    if i.text == ' ':
        attribute.append('-')
    else:
        attribute.append(i.text)
        
for i in driver.find_elements(By.XPATH, '//table[@border="1"]/tbody/tr/td[7]/p')[1:]:
    if i.text == ' ':
        year.append('-')
    else:
        year.append(i.text)
```

```python
ml_df = pd.DataFrame({'Dataset name': dataset_name, 'Data type': data_type, 'Task': task,
                     'Attribute type': attribute_type, 'No. of instances': instances,
                     'No. of attribute': attribute, 'Year': year})
```

```python
ml_df
```

```python
driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/button').click()
```

```python
name=[]
designation=[]
company=[]
skills=[]
location=[]
```

```python
for i in driver.find_elements(By.XPATH, '//div[@id="tabP-1"]/div/div/div[1]/div[1]/p/a[1]'):
    name.append(i.text)

for i in driver.find_elements(By.XPATH, '//div[@id="tabP-1"]/div/div/div[1]/div[1]/p/span[1]'):
    designation.append(i.text)

for i in driver.find_elements(By.XPATH, '//div[@id="tabP-1"]/div/div/div[1]/div[1]/p/a[2]'):
    company.append(i.text)

for i in driver.find_elements(By.XPATH, '//div[@id="tabP-1"]/div/div/div[1]/div[2]'):
    skills.append(i.text)

information_tag = driver.find_elements(By.XPATH, '//div[@id="tabP-1"]/div/div/div[1]/div[1]/p')
for info in information_tag:
    if len(info.find_elements(By.XPATH, './/small[@class="ellipsis"]'))>0:
        for i in info.find_elements(By.XPATH, './/small[@class="ellipsis"]'):
            location.append(i.text)
    else:
        location.append('-')
```

```python
ds_df = pd.DataFrame({'Name': name, 'Designation': designation, 'Company': company,
                      'Skills they hire for': skills, 'Location': location})
```
